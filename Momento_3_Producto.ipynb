# ***ANALISIS DEL DFPETROL***

```python
petrol = pd.read_csv('carros petroleo.csv', sep=';')
```

```python
petrol.head(20)
```

```python
list(petrol['Fuel_Type'].drop_duplicates())
```

```python
sel_petrol = petrol['Fuel_Type'] == 'Petrol'
Npetrolcar = petrol[sel_petrol].shape[0]
Npetrolcar
```

```python
sel_EP = (petrol['Fuel_Type']=='Electricity / Petrol')|(petrol['Fuel_Type']=='Petrol Electric')
NEP= petrol[sel_EP].shape[0]
NEP
```

```python
sel_ECxCost = (petrol['Engine_Capacity'] <= 1300)|(petrol['Fuel_Cost'] != '£2,096')
NECXCOSST = petrol[sel_ECxCost].shape[0]
NECXCOSST
```

```python
print(f"La cantidad de autos que usa petroleo como combustinle encontrados es de: {Npetrolcar}")
print(f"La cantidad de autos con combustible petroleo y electrico es de: {NEP}")
print(f"La cantidad de autos con capacidad menor igual a 1300 y con costo diferente a 2,096 es de: {NECXCOSST}")
```

# ***ANALÍTICA***

```python
petrol['Noise_Level'].mean().round(2)
```

```python
list(petrol['Model'].drop_duplicates())
```

```python
modelos = ['XC90, MY23', 'GT86', 'S-Cross', 'Ignis', 'Ibiza']
sel = petrol['Model'].isin(modelos)
petrol = petrol[sel]
```

```python
petrol['Model'].drop_duplicates()
```

```python
grupo_modelos = petrol.groupby('Model')
grupo_modelos.groups
```

```python
for modelos, data in grupo_modelos:
  print('{}->{}'.format(modelos, data.Noise_Level.mean().round(2)))
```

```python
grupo_modelos[['Fuel_Cost', 'Total_Cost', 'Noise_Level']].mean().round(2)
```
